# Ejercicio 04

Crear un script con **n** filas donde se muestre la siguientes estructura.

```
1

1 2 1

1 3 3 1

1 4 6 4 1

.....
```

````python
#@title Ingrese el número de filas (n)
n_rows = 5 #@param {type: "integer"}
````

````python
def generate_pascals_triangle(num_rows):
    triangle = []
    for i in range(num_rows):
        #longitud de la fila actual (i + 1) y lo inicializamos con tantos 1 como elementos tenga la fila
        row = [1] * (i + 1)
        #print('[',i,']', '*', i+1, ' = ',row)
        #solo calculamos de la 3er fila en adelante
        if i > 1:
            #print('i =',i)
            for j in range(1, i):#desde le 1 hasta la i y el range no incluye e
                #print('j =',j)
                #print("elemento de encima a la izquierda fila anterior", triangle[i-1][j-1])
                #print("elemento de encima a la derecha fila anterior ", triangle[i-1][j])
                #print("row[j]", row[j])
                row[j] = triangle[i-1][j-1] + triangle[i-1][j]
                #print("row[j]", row[j])
        triangle.append(row)
    return triangle

if n_rows <= 0:
    print("Por favor, ingrese un número de filas mayor a 0.")
else:
    pascals_triangle = generate_pascals_triangle(n_rows)
    for row in pascals_triangle:
        print(*row) # spred *
````

#Ejercicio 05

Crear un script para verificar si un texto es palíndromo o no. Para esta comprobación no se deben tener en cuenta los espacios en blanco.

**Es palíndromo:**
```
eva usaba rimel y le miraba suave
```


**No es palíndromo:**
```
puro chile es tu cielo azulado
```



````python
#@title Ingrese el texto a verificar
text_input = "eva usaba rimel y le miraba suave" #@param {type:"string"}
````

````python
def is_palindrome(text):
    # Eliminar espacios y convertir a minúsculas
    cleaned_text = text.replace(" ", "").lower()
    # Comparar el texto limpio con su versión invertida
    return cleaned_text == cleaned_text[::-1]

if is_palindrome(text_input):
    print(f'El texto "{text_input}" es un palíndromo.')
else:
    print(f'El texto "{text_input}" NO es un palíndromo.')
````
