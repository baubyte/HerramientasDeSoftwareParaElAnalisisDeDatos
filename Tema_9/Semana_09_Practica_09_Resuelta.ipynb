# Ejercicio 04

Leer el dataset con padas y a partir de la columna `date` crear 3 nuevas columnas `yea`, `month` y `day` con los valores correspondiente al año, mes y día de la fecha del partido. Sobreescribir el dataset actual.

Hacer los pasos necesarios para que quede traceado por dvc y git, realizar los pushes necesarios.

```python
from pandas import DataFrame, Series, read_csv, concat
```

```python
# leemos el dataset y mostramos la cabecera
df_wc = read_csv(file_name)
print(df_wc.head())
def split_dates(cell:str):
  """ Realiza un split por el caracter -

  Args:
    cell (str): Fecha.

  Return:
    Series: Retorna un objeto Series con las columnas 'year', 'month' y 'day'/
  """
  date  = cell.split("-")
  return Series(data={"year":date[0], "month":date[1], "day":date[2]})
# realizo el apply y el concat de los valores
df_wc = concat((df_wc, df_wc["date"].apply(split_dates)), axis=1)
# muestro el nuevo DataFrame
print()
print(df_wc.head())
# guardo el archivo
df_wc.to_csv(file_name)
```

# Ejercicio 06

Deshacer el último cambio del ejercicio 05.

Hacer los pasos necesarios para que quede traceado por dvc y git, realizar los pushes necesarios.

```python
# vamos a ver el log de git y guardar los 2 últimos commit
!git log
```

```python
# vamos a guardar el checksum del ante penúltimo commit en un archivo dentro del directorio tmp
!git log | grep commit | head -n 2 | tail -n 1 | awk '{print $2}' > /tmp/second_commit
```

```python
# vamos a mostrar el contenido de los archivo que generamos
!cat /tmp/second_commit
```

```python
# primero hacemos un checkout de git
!echo "La sentencia ejecutada es:"
# los archivos contienen los checksum para que se pueda replicar en distintas
# ejecuciones
!echo "!git checkout $(cat /tmp/second_commit)"
!echo ""
!echo ""

!git checkout $(cat /tmp/second_commit)
```

```python
# estado del proyecto de git
!git status
# como podemos ver en este caso tenemos un detached porque el head se desplazó
# del último commit
```

```python
# mostramos las primeras 5 líneas del archivo data/wc_matches.csv
!head -n 5 {file_name}
```

```python
# hacemos un checkout de dvc
!dvc checkout
# vemos que recuperamos el archivo con la versión anterior
```

```python
# mostramos las primeras 5 líneas del archivo data/wc_matches.csv
!head -n 5 {file_name}
# como podemos ver, este archivo no posee la columna total_score
```

```python
# estado del proyecto de dvc
!dvc status
# nos muestra que no hay cambios
```

```python
# estado del proyecto de git
!git status
# seguimos con el desplazamiento
```

```python
# como queremos conservar los cambios vamos a crear un rama temporal
!git branch temporal
# nos movemos a la rama master
!git checkout main
# hacemos un merge entre la rama master y la temporal
!git merge temporal
# borramos la rama temporal
!git branch -d temporal
# estado del proyecto de git
!git status
```

```python
# recuperamos el archivo y estamos sin el desplazamiento o detached
!head -n 5 {file_name}
```

```python
# estado del proyecto de dvc
!dvc status
```

```python
# agregamos a dvc
!dvc add {file_name}
```

```python
# agregamos el archivo recuperado
!git add {file_name_dvc}
```
