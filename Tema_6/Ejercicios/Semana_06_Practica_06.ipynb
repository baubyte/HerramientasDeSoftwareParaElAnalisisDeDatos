```python
import os
import pandas as pd
from google.colab import userdata, drive, files
import re
from tabulate import tabulate
import nltk
from nltk.corpus import stopwords
import spacy
from nltk.stem import SnowballStemmer
```

# Ejercicio 02

Al dataset anterior agregarle una nueva columna llamada `nltk_sin_stopword` quitando los stop words de la columna `news`según nltk.


```python
nltk.download('stopwords')
spanish_stopwords = set(stopwords.words('spanish'))

def remove_nltk_stopwords(text):
    if isinstance(text, str):
        words = re.findall(r'\b\w+\b', text.lower())
        filtered_words = [word for word in words if word not in spanish_stopwords]
        return ' '.join(filtered_words)
    return text

df_spanish_news_classificatio['nltk_sin_stopword'] = [
    remove_nltk_stopwords(text) for text in df_spanish_news_classificatio['news'].values
]

display(df_spanish_news_classificatio[['news', 'nltk_sin_stopword']].head())
```

# Ejercicio 03

Al dataset anterior agregarle una nueva columna llamada `nltk_stemn_sin_stopword` cambiando las palabras de la columna `nltk_sin_stopword` por su stem según nltk.


```python
spanish_stemmer = SnowballStemmer('spanish')

def stem_text(text):
    if isinstance(text, str):
        words = re.findall(r'\b\w+\b', text.lower())
        stemmed_words = [spanish_stemmer.stem(word) for word in words]
        return ' '.join(stemmed_words)
    return text

df_spanish_news_classificatio['nltk_stemn_sin_stopword'] = [
    stem_text(text) for text in df_spanish_news_classificatio['nltk_sin_stopword'].values
]

# Mostrar el resultado
display(df_spanish_news_classificatio[['nltk_sin_stopword', 'nltk_stemn_sin_stopword']].head())
```

# Ejercicio 04

Al dataset anterior agregarle una nueva columna llamada `spacy_sin_stopword` quitando los stop words de la columna `news` según spacy.


```python
nlp_es_sm = spacy.load('es_core_news_sm')

def remove_spacy_stopwords(text):
    if isinstance(text, str):
        doc = nlp_es_sm(text.lower())
        filtered_words = [token.text for token in doc if not token.is_stop and not token.is_punct and not token.is_space]
        return ' '.join(filtered_words)
    return text

df_spanish_news_classificatio['spacy_sin_stopword'] = [
    remove_spacy_stopwords(text) for text in df_spanish_news_classificatio['news'].values
]

# Mostrar el resultado
display(df_spanish_news_classificatio[['news', 'spacy_sin_stopword']].head())
```

# Ejercicio 05

Al dataset anterior agregarle una nueva columna llamada `spacy_lemma_sin_stopword` cambiando las palabras de la columna `spacy_sin_stopword` por su lemma según spacy.


```python
nlp_es_sm = spacy.load('es_core_news_sm')

def lemmatize_spacy_text(text):
    if isinstance(text, str):
        doc = nlp_es_sm(text)
        lemmas = [token.lemma_ for token in doc]
        return ' '.join(lemmas)
    return text

df_spanish_news_classificatio['spacy_lemma_sin_stopword'] = [
    lemmatize_spacy_text(text) for text in df_spanish_news_classificatio['spacy_sin_stopword'].values
]

display(df_spanish_news_classificatio[['spacy_sin_stopword', 'spacy_lemma_sin_stopword']].head())
```
