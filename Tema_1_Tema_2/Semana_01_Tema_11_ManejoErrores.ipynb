# Manejo de errores

```python
try:
  # Sentencias
except Exception as e:
  # Sentencias
finally:
  # Sentencias
```

````python
valor:int = 1
muestra:str = "valor: " + valor
print(muestra)
````

````python
valor:int = 1
try:
  muestra:str = "valor: " + valor
except TypeError as e:
  print("No se puede concatenar un string y un entero.")
  print("Conversión automática del valor.")
  muestra:str = "valor: " + str(valor)
print(muestra)
````

````python
valor:int = 1
muestra:str = "valor: "
try:
  muestra += valor
except TypeError as e:
  pass
print(muestra)
````

````python
valor:int = 1
muestra:str = "valor: "
try:
  muestra += "try"
  muestra += valor
except TypeError as e:
  muestra += "_except"
finally:
  muestra += "!"
print(muestra)
````

````python
valor:int = 1
muestra:str = "valor: "
if isinstance(valor, int):
  raise TypeError(100, "El valor no puede ser entero.")
try:
  muestra += valor
except TypeError as e:
  pass
finally:
  muestra += "!"
print(muestra)
````
