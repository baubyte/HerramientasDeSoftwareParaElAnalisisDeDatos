# Ejercicio 01

Crear un script que lea el archivo `/content/sample_data/README.md` y cuenta la cantidad de letras. Tener en cuenta que hay letras que están en mayúscula y otras en minúsculas.

Mostrar una tabla con la letra y la cantidad de veces que se repite. No se deben mostrar letras que no existan en el texto. 

```python
! pip install tabulate
```

```python
from tabulate import tabulate
file_md: str = "/content/sample_data/README.md"

def count_characters(text: str) -> dict:
    char_counts = {}
    for char in text:
        keyChar = char
        if char == "\n":
            keyChar = "enter"
        elif char == " ":
            keyChar = "space"
        elif char == "\t":
            keyChar = "tab"
        char_counts[keyChar] = char_counts.get(keyChar, 0) + 1
    return char_counts

with open(file_md, "r") as file:
    content = file.read()

character_repetition = count_characters(content)
print(tabulate(sorted(character_repetition.items()), headers=["Character", "Count"]))
```

# Ejercicio 02

Crear un script que lea el archivo `/content/sample_data/anscombe.json` y guarde en un nuevo archivo los objetos que se encuentran en las posiciones pares.

```python
from json import load, dump

file_json: str = "/content/sample_data/anscombe.json"
file_json_even: str = "/content/sample_data/anscombe_even.json"

with open(file_json, "r") as file:
    json_data = load(file)

json_data_even = [obj for position, obj in enumerate(json_data) if position % 2 == 0]

with open(file_json_even, "w") as file:
    dump(json_data_even, file, indent=4)
```

# Ejercicio 03

Crear un script que lea el archivo `/content/sample_data/california_housing_test.csv` y genere un nuevo archivo donde la columna `housing_median_age` es igual a 30.

```python
from csv import reader, writer

file_csv: str = "/content/sample_data/california_housing_test.csv"
file_csv_median_age: str = "/content/sample_data/california_housing_median_income_test.csv"

with open(file_csv, "r") as file:
    read_csv = reader(file)
    list_csv = list(read_csv)

head = list_csv.pop(0)
data_median_age = [row for row in list_csv if float(row[head.index("housing_median_age")]) == 30.0]

with open(file_csv_median_age, "w") as file:
    write_csv = writer(file)
    write_csv.writerow(head)
    write_csv.writerows(data_median_age)
```

# Ejercicio 04

Crear un script que lea el archivo `/content/sample_data/mnist_test.csv` y genere un nuevo archivo donde la fila seleccionada para grabar debe poseer el número 253 solo 3 veces en toda la fila.

```python
from csv import reader, writer

file_minist_csv = "/content/sample_data/mnist_test.csv"
file_minist_253_csv = "/content/sample_data/mnist_253.csv"

with open(file_minist_csv, "r") as file_mnist:
    read_mnist = reader(file_mnist)
    list_mnist = list(read_mnist)

head = list_mnist.pop(0)

list_mnist_253 = [row for row in list_mnist if row.count("253") == 3]

with open(file_minist_253_csv, "w") as file_mnist_253:
    write_mnist_253 = writer(file_mnist_253)
    write_mnist_253.writerow(head)
    write_mnist_253.writerows(list_mnist_253)
```

# Ejercicio 05

Crear un script que lea el archivo `/content/sample_data/anscombe.json`, transformarlo a csv, ini, yml y guardarlos en nuevos archivos.


```python
! pip install pyyaml
```

```python
from json import load, dump as json_dump
from csv import reader, writer
from configparser import ConfigParser
from yaml import dump as yaml_dump

file_json: str = "/content/sample_data/anscombe.json"
file_csv: str = "/content/sample_data/anscombe.csv"
file_ini: str = "/content/sample_data/anscombe.ini"
file_yml: str = "/content/sample_data/anscombe.yml"

json_data: dict = {}

with open(file_json, "r") as file:
    json_data = load(file)
#[dict_values(['I', 10.0, 8.04]), dict_values(['I', 8.0, 6.95])]

# Transformar a CSV
headers = list(json_data[0].keys())
rows = [list(data_row.values()) for data_row in json_data]
with open(file_csv, "w") as file:
    csv_writer = writer(file)
    csv_writer.writerow(headers)
    csv_writer.writerows(rows)

# Transformar a INI
config = ConfigParser()
for index, data in enumerate(json_data):
    section_name = f"row_{index}"
    config[section_name] = {str(key): str(value) for key, value in data.items()}
print(config)
with open(file_ini, "w", encoding="utf-8") as file:
    config.write(file)

# Transformar a YAML
yaml_data: dict = {}
for index, data in enumerate(json_data):
    section_name: str = f"row_{index}"
    yaml_data.update({section_name: data})
with open(file_yml, "w", encoding="utf-8") as file:
    yaml_dump(yaml_data, file)
```
