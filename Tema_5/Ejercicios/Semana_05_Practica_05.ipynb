```python
from bs4 import BeautifulSoup
import re
import requests
import os

def descarga_dataset(url:str, nombre_del_archivo: str):
  """ Descarga el archivo de normas legales. Utilizar esta función porque
      el archivo es demasiado grande y muestra una pantalla de acuerdo.

      Args:
        url (str): La URL del archivo en Google Drive.
        nombre_del_archivo (str): Contiene el nombre del archivo con el que se
          guardará en el disco rígido.
  """
  matches = re.findall(r'/file/d/([a-zA-Z0-9_-]+)', url)

  if matches:
    file_id = matches[0]
    print("File ID:", file_id)
    url_download = f"https://drive.google.com/uc?id={file_id}"
    try:
      response = requests.get(url=url_download)
      page = BeautifulSoup(response.content, "html.parser")
      url = f'"{page.find("form", id="download-form").get("action")}"'
    except Exception:
      url = url_download

    # Especificar el path de descarga
    path_descarga = f'/content/{nombre_del_archivo}'

    # Crear directorio si no existe
    os.makedirs(os.path.dirname(path_descarga), exist_ok=True)

    # Descargar el archivo en el path especificado
    !wget -X POST {url} -O {path_descarga}
  else:
    print("File ID not found")
```
