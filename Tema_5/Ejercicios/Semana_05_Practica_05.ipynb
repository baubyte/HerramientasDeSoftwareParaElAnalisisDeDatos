```python
from bs4 import BeautifulSoup
import re
import requests
import os

def descarga_dataset(url:str, nombre_del_archivo: str):
  """ Descarga el archivo de normas legales. Utilizar esta función porque
      el archivo es demasiado grande y muestra una pantalla de acuerdo.

      Args:
        url (str): La URL del archivo en Google Drive.
        nombre_del_archivo (str): Contiene el nombre del archivo con el que se
          guardará en el disco rígido.
  """
  matches = re.findall(r'/file/d/([a-zA-Z0-9_-]+)', url)

  if matches:
    file_id = matches[0]
    print("File ID:", file_id)
    url_download = f"https://drive.google.com/uc?id={file_id}"
    try:
      response = requests.get(url=url_download)
      page = BeautifulSoup(response.content, "html.parser")
      url = f'"{page.find("form", id="download-form").get("action")}"'
    except Exception:
      url = url_download

    # Especificar el path de descarga
    path_descarga = f'/content/{nombre_del_archivo}'

    # Crear directorio si no existe
    os.makedirs(os.path.dirname(path_descarga), exist_ok=True)

    # Descargar el archivo en el path especificado
    !wget -X POST {url} -O {path_descarga}
  else:
    print("File ID not found")
```

# Ejercicio 03

Obtener la mayor cantidad de goles en un partido que hizo cada selección. Mostrar El nombre del país de la selección y la cantidad de goles.

```python
df_scores_team1 = df_wc_matches.groupby('team1').aggregate({'score1': 'max'}).rename(columns={'score1': 'max_score'})
df_scores_team2 = df_wc_matches.groupby('team2').aggregate({'score2': 'max'}).rename(columns={'score2': 'max_score'})
df_scores_teams_concat = pd.concat([df_scores_team1, df_scores_team2])
df_teams_scores = df_scores_teams_concat.groupby(df_scores_teams_concat.index)['max_score'].max()
df_teams_scores = df_teams_scores.reset_index()
df_teams_scores.columns = ['team', 'max_goals']
print("Paises con mayor cantidad de goles:")
display(df_teams_scores.sort_values(by='max_goals', ascending=False))
```

Levantar el siguiente dataset de kaggle `stephenknoth/spanish-language-articles` en un DataFrame donde cada fila sea un texto/archivo.

`El directorio tiene varios archivos.`

`Los archivos deberían dar un error al leerlos, ver el tema de encoding.`

```python
!pip install -q kaggle
dataset_name = "stephenknoth/spanish-language-articles"
dir_dataset_spanish = f"{dir_dataset}/spanish-language-articles"
os.makedirs(dir_dataset_spanish, exist_ok=True)
!kaggle datasets download -d {dataset_name} -p {dir_dataset_spanish} --unzip
```

```python
list_files: list = os.listdir(dir_dataset_spanish)
list_data_files: list = []

for file in list_files:
    with open(f"{dir_dataset_spanish}/{file}", "r", encoding="latin1") as f:
        print(f"Leyendo archivo (latin1): {file}")
        text = f.read()
        list_data_files.append({"filename": file, "content": text})
```

```python
df_articles = pd.DataFrame(list_data_files)
print(f"DataFrame de artículos creado con {len(df_articles)} filas.")
display(df_articles.head())
```
