# Listas


```python
# list_:list=list(1,2,3,4)
list_:list = [1,2,3,4]
print(f"list_: {list_} - tipo: {type(list_)}")
```

```python
# append
list_.append(5)
print(f"list_: {list_}")
```

```python
# insert
list_.insert(0, 0)
print(f"list_: {list_}")
```

```python
# extend
list_2 = [6, 7]
# list_ = list_ + list_2
list_.extend(list_2)
print(f"list_: {list_}")
```

```python
# count
list_.append(7)
print(f"list_: {list_}")
print(f"count (7): {list_.count(7)}")
```

```python
# index
print(f"list_: {list_}")
print(f"index(5): {list_.index(5)}")
print(f"index(7): {list_.index(7)}")
```

```python
# remove
list_.remove(6)
print(f"list_: {list_}")
```

```python
# pop
list_pop = list_.pop()
print(f"list_: {list_}")
print(f"list_pop: {list_pop}")
```

```python
# len
print(f"list_: {list_}")
print(f"list_: {len(list_)}")
```

```python
# clear
list_.clear()
print(f"list_: {list_}")
print(f"list_: {len(list_)}")
```

# Set

```python
#set_:set=set([1,2,3,4])
set_:set={1,2,3,4,3,4,5,3}
print(f"set_: {set_}")
```

```python
# add
set_.add(4)
print(f"set_: {set_}")
set_.add(6)
print(f"set_: {set_}")
```

```python
# update
set_.update({1,2,7})
print(f"set_: {set_}")
```

```python
# remove
set_.remove(6)
print(f"set_: {set_}")
```

```python
# pop
set_valor=set_.pop()
print(f"set_: {set_}")
print(f"set_valor: {set_valor}")
```

```python
# len
print(f"set_: {set_}")
print(f"set_: {len(set_)}")

```

```python
# clear
set_.clear()
print(f"set_: {set_}")
print(f"set_: {len(set_)}")
```

# Tupla

# Diccionario

```python
#dict_:dict=dict()
dict_:dict={"1": "a","2": "b","3": "c"}
print(f"dict_: {dict_}")
```

```python
# []
dict_["4"] = "d"
print(f"dict_: {dict_}")
dict_["3"] = "cd"
print(f"dict_: {dict_}")
dict_["3"] = "c"
print(f"dict_: {dict_}")
```

```python
# update
print(f"dict_: {dict_}")
dict_.update({"4": "df", "5": "f"})
print(f"dict_: {dict_}")
dict_["4"] = "d"
print(f"dict_: {dict_}")
```

```python
# pop
print(f"dict_: {dict_}")
print(f"pop: {dict_.pop('5')}")
print(f"dict_: {dict_}")
```

```python
# keys
print(f"keys: {list(dict_.keys())}")
```

```python
# values
print(f"keys: {list(dict_.values())}")
```

```python
# items
for key,value in dict_.items():
  print(f"key: {key} - value: {value}")
```

```python
# len
print(f"dict_: {dict_}")
print(f"len: {len(dict_)}")
```

```python
# clear
dict_.clear()
print(f"dict_: {dict_}")
print(f"len: {len(dict_)}")
```
